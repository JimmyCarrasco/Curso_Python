Tarea 1

Con los datos presentados en el archivo Metales.xlsx (el cual presenta la cantidad extraída de ciertos metales por año en el Perú), responda las siguientes preguntas:

1. ¿Cuál es el promedio de oro extraído entre los años 2000-2022?
2. ¿Cuánto es la cantidad máxima de plata extraída en un año y en qué año ocurrió?
3. ¿Cuánto es la cantidad mínima de cobre extraída en un año y en qué año ocurrió?
4. ¿Cuánto es la cantidad total de cobre extraído entre los años 2000-2022?
5. ¿Cuál es el promedio de oro extraído entre los años 2010-2022?





1. ¿Cuál es el promedio de oro extraído entre los años 2000-2022?

```python
data = pd.read_excel('Metales.xlsx')
data.head()
```

```python
data = data.set_index("AÑO")
data = data.loc[2020:2022,['ORO(KG)']]
print("EL promedio de Oro estraido es: ", data.mean().tolist())
```

2. ¿Cuánto es la cantidad máxima de plata extraída en un año y en qué año ocurrió?


```python
data2 = pd.read_excel('Metales.xlsx')
data2.head()
```

```python
maximo = data2['PLATA(KG)'].max()
print("La cantidad maxima de PLATA en KG es :  " ,maximo)
data2 = data2[data2['PLATA(KG)'] == maximo ]
fecha_max = data2["AÑO"].tolist()
print('La fechja en que ocurrio el maximo es: ',fecha_max)
```

3. ¿Cuánto es la cantidad mínima de cobre extraída en un año y en qué año ocurrió?


```python
data3 = pd.read_excel('Metales.xlsx')
data3.head()
```

```python
minimo = data3['COBRE(TMF)'].min()
print("La cantidad minima de cobre en TMF es :  " ,minimo)
data3 = data3[data3['COBRE(TMF)'] == minimo]
fecha_min = data3['AÑO'].tolist()
print('La fecha en que ocurrio el minimo es: ',fecha_min)
```

4. ¿Cuánto es la cantidad total de cobre extraído entre los años 2000-2022?


```python
data4 = pd.read_excel('Metales.xlsx')
data4.head()
```

```python
data4 = data4.set_index("AÑO")
data4 = data4.loc[2000:2022,['COBRE(TMF)']]
print("EL promedio de Cobre estraido es: ", data4.sum().tolist())
```

5. ¿Cuál es el promedio de oro extraído entre los años 2010-2022?

```python
data5 = pd.read_excel('Metales.xlsx')
data5.head()
```

```python
data5 = data5.set_index('AÑO')
data5 = data5.loc[2010:2022,['ORO(KG)']]
print("EL promedio de Oro estraido es: ", data5.mean().tolist())
```
