## ¿Qué es la recursión?

Imágina que quieres hallar el factorial de 4:

$n!=\begin{pmatrix}1; \;\;\;\;  n==0 \\ 1; \;\;\;\;   n==1 \\ n*(n-1) ; \;\;\;\; n>1 \end{pmatrix}$   
    
    Si n = 4, entonces:
        n = n * (3)!, pero:
        3! = 3 * 2!, pero:
        2! = 2 * 1!, pero:
        1! = 1, retorna 1.

Luego:

    2! = 2
    3! = 3 * 2! = 6
    4! = 4 * 3! = 24

Como se puede ver, la función n! se llama a sí misma varias veces. Esto es una función recursiva, es decir, una función que se llama a sí misma hasta que llega a un punto de parada. Este punto de parada se llama caso base.

En el ejemplo anterior, el punto de parada es cuando n=1.   

---------------------------------


Dos cosas que se deben identificar en una función recursiva son:

a) **Caso Base:**  Es el punto de parada de la recursión.   
b) **Algoritmo:** El conjunto de instrucciones que ejecuta el programa para resolver el problema.

--------------------------

```python
#1. Factorial

def fac(n):
  # Caso base
  if (n==0 or n==1):
    return 1
  # Cuerpo del algoritmo
  return n*fac(n-1)


def pow_2(n):
  if(n==0):
    return 1
  if(n==1):
    return 2
  return 2*pow_2(n-1)

print(pow_2(4))

def ejer2(n):
  if(n==1):
    return 2
  return 2**n + ejer2(n-1)

x = [1,2,3,4,5,6]
print("head")

print(x[0])

print("tail")

print(x[1:])

def fib(n):
  if(n==1 or n==2):
    return 1
  return fib(n-1) + fib(n-2)

print(fib(6))


def find_sec_r(lista,e):
   if(lista == []):
    return False
   if(lista[0] == e):
    return True

   return find_sec_r(lista[1:],e)

l = [6,5,4,3]

print(find_sec_r(l,8))

def mayor(lista):
  if(len(lista) == 1):
    return lista[0]
  return max(lista[0], mayor(lista[1:]))

```
