## Búsqueda Binaria
Imagina que tienes un cuaderno de 500 hojas. En cada hoja hay una palabra escrita. Todas las palabras escritas en el cuaderno son completamente diferentes y están ordenadas alfabéticamente. Esto significa que las palabras que inician con A están antes que las que inician con B y así sucesivamente.

Mejor caso se da cuanto el elemento que se busca está justo en la mitad es decir se demora $\Omega(1)$   
El peor caso se da cuanto el elemento no está y se demora $log(n)$

```python
# Programar
def busqueda_binaria(libros,x):
  ini = 0
  fin = len(libros) - 1
  while(ini<=fin):
    mitad = (ini + fin)//2
    if (libros[mitad]==x):
      return True
    if (libros[mitad]>x):
      fin = mitad - 1
    else:
      ini = mitad + 1
  return False

print(busqueda_binaria([2,4,7,12,14,15,17,18],149))



```

## Buble sort.

burbuja(cartas, n)   
Inicio
> Para i=0 hasta llegar a  n      
  - Para j=0 hasta llegar hasta (n-1)
    -  Si (cartas[j]>cartas[j+1])   
     - mano = cartas[j]
     - cartas[j] =  cartas[j+1]   
     - cartas[j+1] = mano    

Fin   

Ver video: [Bublesort](https://www.youtube.com/watch?v=Iv3vgjM8Pv4)





```python
# Implementar código
def burbuja(vector):
  for j in range(len(vector)):
    for i in range(0,len(vector) -1-j):
      if(vector[i]>vector[i+1]):
        tmp = vector[i]
        vector[i] = vector[i+1]
        vector[i+1] = tmp
  return vector

print(burbuja([6,5,14,3,2,1]))
```

## Select Sort

```python
# Algoritmo
def find_min(v,pos):
  menor = v[pos]
  ind = pos
  for i in range(pos+1,len(v)):
     if(v[i]<menor):
       menor = v[i]
       ind = i
  return menor, ind

def select_sort(v):
  for i in range(len(v)):
    menor, ind = find_min(v,i)
    tmp = v[ind]
    v[ind] = v[i]
    v[i]=tmp
  return v

print(select_sort([9,8,7,6,5,4,3,2,1]))



```
